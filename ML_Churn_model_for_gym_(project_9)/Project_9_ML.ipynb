# Моделирование оттока клиентов для фитнес-клуба

## Описание задачи

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.

Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Как понять, что клиент больше не с вами? Можно записать в отток тех, кто попросил закрыть договор или удалил аккаунт. Однако клиенты не всегда уходят демонстративно: чаще перестают пользоваться сервисом тихо.

Индикаторы оттока зависят от специфики отрасли. Когда пользователь редко, но стабильно закупается в интернет-магазине — не похоже, что он «отвалился». А вот если две недели не заходит на канал с ежедневно обновляемым контентом, дела плохи: подписчик заскучал и, кажется, оставил вас.

Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Конечно, не исключено, что он уехал на Бали и по приезде обязательно продолжит ходить на фитнес. Однако чаще бывает наоборот. Если клиент начал новую жизнь с понедельника, немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.

Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Ваша __задача — провести анализ и подготовить план действий по удержанию клиентов__.

А именно:

- научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
- сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
- проанализировать основные признаки, наиболее сильно влияющие на отток;
- сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
    1. выделить целевые группы клиентов;
    2. предложить меры по снижению оттока;
    3. определить другие особенности взаимодействия с клиентами.

## Описание данных

Набор данных включает следующие поля:
- `'Churn'` — факт оттока в текущем месяце;
- Текущие поля в датасете:
    - Данные клиента за предыдущий до проверки факта оттока месяц:
        - `'gender'` — пол;
        - `'Near_Location'` — проживание или работа в районе, где находится фитнес-центр;
        - `'Partner'` — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
        - `'Promo_friends'` — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
        - `'Phone'` — наличие контактного телефона;
        - `'Age'` — возраст;
        - `'Lifetime'` — время с момента первого обращения в фитнес-центр (в месяцах).
- Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
    - `'Contract_period'` — длительность текущего действующего абонемента (месяц, 3 месяца, 6 месяцев, год);
    - `'Month_to_end_contract'` — срок до окончания текущего действующего абонемента (в месяцах);
    - `'Group_visits'` — факт посещения групповых занятий;
    - `'Avg_class_frequency_total'` — средняя частота посещений в неделю за все время с начала действия абонемента;
    - `'Avg_class_frequency_current_month'` — средняя частота посещений в неделю за предыдущий месяц;
    - `'Avg_additional_charges_total'` — суммарная выручка от других услуг фитнес-центра: кафе, спорт-товары, косметический и массажный салон.

# Модель оттока

## Подготовка выборки

```python
X = df.drop('churn', axis=1)
y = df['churn']
```

Делаем разбивку на обучающую и валидационную выборки:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Cтандартизируем данные:

```python
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train) 
X_test_st = scaler.transform(X_test)
```

## Обучаем модели

```python
models = [LogisticRegression(random_state=0), RandomForestClassifier(random_state=0)]
```

```python
#Функция на вход получает название модели, обучающую выборку 
#из матрицы и вектора ответов, а также валидирующую матрицу и вектор ответов.
#Обучает модель и строит вектор предсказаний по валидационной выборке.
#На выходе печатает основные метрики оценки качества модели
def make_prediction(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('Accuracy: {:.2f}'.format(accuracy_score(y_test, y_pred)))
    print('Precision: {:.2f}'.format(precision_score (y_test, y_pred)))
    print('Recall: {:.2f}'.format(recall_score (y_test, y_pred)))
```

```python
for m in models:
    print(m)
    make_prediction(m, X_train_st, y_train, X_test_st, y_test)
    print('')
    print('-----------------------------------------------------------')
    print('')
```

## Вывод по моделям

Применили две модели машинного обучения к нашим признакам: логистическую регрессию и случайный лес. Предварительно мы стандартизировали данные.

Видим, что логистическая регрессия оказалась лучше по всем трем ключевым метрикам, в особенности, по очень важной для нас метрике `recall`, поскольку мы можем позволить себе ошибочно определить лояльного клиента, как потенциального "беглеца", но очень не хотим проглядеть настоящих уходящих клиентов.

Интересно, кстати посмотреть на распределение весов:

```python
model = models[0]
model.fit(X_train_st, y_train)

dictionary = dict(zip(X.columns.tolist(), model.coef_.tolist()[0]))
pd.DataFrame(data = dictionary, index=['weight']).T.round(2).style.apply(lambda x: ["background: red" if abs(v) >= 1.0 else "" for v in x], axis = 0)
```

Наибольшее влияние на модель оказали параметры `contract_period`, `age` и `life_time`, последний особенно.

Это вполне логично - старшие пользователи с длинными контрактами, которые уже давно пользуются услугами, просто обязаны быть "золотым фондом" и опорой бизнеса.

# Класстеризация клиентов

## Матрица расстояний и дендрограмма

Стадратизируем наши данные и строим матрицу расстояний

```python
X_sc = scaler.fit_transform(X)

linked = linkage(X_sc, method = 'ward') 
```

Нарисуем по нашей матрице дендрограмму:

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Дендрограмма матрицы расстояний для признаков')
plt.show() 
```

Из дендрограмы видим, что оптимально будет использовать разбить наших клиентов на __4 кластера__.

## K-Means

### Распределение средних по кластерам

Теперь применим алгоритм K-means c 5 кластерами (тут стоило бы взять 4 кластера, раз уж мы получили этот результат на предыдущем шаге, но в силу требований учебного проекта и для унификации возьмем 5) к нашей выборке:

```python
km = KMeans(n_clusters = 5)
labels = km.fit_predict(X_sc)
df['cluster'] = labels
df.groupby(['cluster']).mean().sort_values(by='age', ascending=True)
```

Более возрастные клиенты больше тратят, заключают более длительные контракты, чаще посещают зал и т.д.

### Распределение признаков по кластерам

Информация из предыдущего раздела подсказывает нам, что логично обозначать кластеры средним возрастом клиентов из них. По убыванию возраста кластеры идут в следующем порядке:

```python
for i in df.groupby('cluster')['age'].mean().sort_values().index:
    print(i)
```

Построим гистограммы распределения частостот наших параметров по кластерам, обозначенным средним возрастом клиента.

```python
#Определяем порядок индексов кластеров по возрастанию среднего возраста
ages_means_cluster_sorted = df.groupby('cluster')['age'].mean().sort_values()
index = ages_means_cluster_sorted.index

for col in df.drop('cluster', axis=1).columns:
    bins = 20
    for i in index:
        sns.distplot(df.query('cluster == @i')[col], kde=False, norm_hist=True, bins=bins)
    #sns.distplot(no_churn[col], kde=False, norm_hist=True, color='green', bins=bins)
    plt.legend(labels = ages_means_cluster_sorted.round(2))
    #plt.legend()
    plt.title(f'Гистограмма частот распределения {col} по кластерам')
    plt.show()
```

В принципе, никакой новой информации мы не получили, все выводы уже сделаны при анализе средних.

Можно отметить довольно большой разброс данных о возрасте внутри каждого кластера. Это говорит о том, что будет трудно судить о принадлежности клиента к тому или иному кластеру, исключительно на основании возраста. Безусловно, стоит рассматривать и другие показатели.
