# Стримчик 2016

## Описание проекта

Мы работаем в **интернет-магазине «Стримчик»**, который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

- **E** & **K-A** - Для всех
- **EC** - Для детей младшего возраста
- **E10+** - Для всех от 10 лет и старше
- **T** - Подросткам
- **M** - Для взрослых
- **AO** - Только для взрослых
- **RP** - Рейтинг ожидается

## Описание данных

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

Видим, что помимо `NaN`'ов есть еще какой-то `tbd`, видимо to be determined, в количестве 2376 штук.

Таким образом, общий процент неизвестных значений в данном столбце целых 54.6%

```python
na_user_score_percent = (df.query('user_score == "tbd"')['user_score'].count() + df['user_score'].isna().sum())/len(df)*100
print(na_user_score_percent)
```

Интересно, есть ли корреляция `critic_score` и `user_score`

```python
#Отбросим NaN и tbd и сделаем конвертацию в float
user_score_not_empty = df[(df['user_score'].notna()) & (df['user_score']!="tbd")]['user_score'].astype('float')
print(user_score_not_empty)
critic_score_not_empty = df[df['critic_score'].notna()]['critic_score']
print(critic_score_not_empty)
```

```python
pearson = critic_score_not_empty.corr(user_score_not_empty, method='pearson')
print(pearson)
spearman = critic_score_not_empty.corr(user_score_not_empty, method='spearman')
print(spearman)
```

Посмотрим на корреляцию рангов и просто по Пирсону с уровнем значимости 5%

```python
# Уберем NaN и tbd, чтобы остались только полные пары.
df1 = df.dropna(subset=['critic_score', 'user_score'])
df1 = df1[df1['user_score']!="tbd"]
```

```python

figure(num=None, figsize=(9, 9), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(df1['critic_score'],df1['user_score'].astype('float'), alpha = 0.2)
plt.title('Корреляция оценок критиков и пользователей')
plt.grid(True,linestyle='--')
plt.show()
```

```python
stat, p = spearmanr(df1['critic_score'], df1['user_score'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	  print('Probably independent')
else:
	  print('Probably dependent')
```

```python
stat, p = pearsonr(df1['critic_score'], df1['user_score'].astype('float'))
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	  print('Probably independent')
else:
	  print('Probably dependent')
```

**Корреляция, как линейная, так и рангов имеется**, хотя и не самая явная. 

Естественно, удалить 50% строк не представляется возможным.

Заменим `tbd` в *user_score* на `NaN` и преобразуем его в `float`

```python
df['user_score'] = df['user_score'].replace("tbd", np.NaN)
df['user_score'] = df['user_score'].astype('float')
#Проверим, что остался тот же процент пропущенных
df['user_score'].isna().sum()/len(df)*100 == na_user_score_percent
```

## Распределение выпуска игр по годам

```python
def bar_plot(data, title):
    plt.bar(data.index, data)
    plt.title(title)
    plt.grid(True)

def plot_plot(data, title):
    plt.plot(data.index, data)
    plt.title(title)
    plt.grid(True)
```

```python
year_count = df['year_of_release'].value_counts(normalize=True).sort_index()*100
print(year_count.round(2))
plot_plot(year_count, 'Процент выпущенных игр по годам')
plt.show()
```

Мне кажется, что игры стали масштабнее и дороже, а мелкие indie проекты перешли на мобильные платформы. Вообще мобильный рынок сильно оттянул спрос на себя, по моему мнению.
    
Большие проекты также стали зарабатывать на DLC, что тоже ведет к сокращению количества новых игр.

Видим, что имеется длинный хвост в ранние годы. Можно отбросить, скажем 5% распределения.

```python
tail_length = 0
for ind in year_count.index:
    if tail_length > 5.:
        break
    tail_length += year_count[ind]
    print('%.2f' % tail_length, ind)
```

Получается можно **отбросить до 1995 включительно**.

## Продажи по платформам

Выберем топ 20 платформ по суммарным продажам

```python
top_platform_by_sales = df.groupby('platform')['total_sales'].sum()\
.sort_values(ascending=False).head(20)
print(top_platform_by_sales)
```

По каждой из них посмотрим на распределение количества игр на этой платформе по годам:

```python
for platform in top_platform_by_sales.index:
    data = df.query('platform == @platform')['year_of_release'].value_counts()
    name = f'Количество игр на {platform}'
    bar_plot(data, name)
    plt.show()
```

```python
first_years = []
last_years = []
for platform in top_platform_by_sales.index:   
    
    first_year = df.query('platform == @platform')['year_of_release'].min()
    last_year = df.query('platform == @platform')['year_of_release'].max()
    first_years.append(first_year)
    last_years.append(last_year)
    
print(first_years) 
print(last_years)
lifetime = [a - b for a, b in zip(last_years, first_years)]

top20_lifetime = pd.DataFrame({'platform':top_platform_by_sales.index.tolist(), 
              'first_year':first_years, 
              'last_year':last_years, 
              'lifetime':lifetime}) 
top20_lifetime
```

```python
top20_lifetime['lifetime'].hist(bins = 30)
plt.title('Распределения длительности жизни для топ 20 платформ')
plt.show()
```

```python
top20_lifetime['lifetime'].describe()
```

Из построенных графиков и расчетов видим, что **цикл жизни платформы где-то от 6 до 11 лет**. При этом есть существенные выбросы в виде PC и DS.

## Равенство средних пользовательских рейтингов XOne и PC 

```python
ALPHA = .05
```

Сформулируем наши гипотезы:
$$
H_0: \mu_x =\mu_{pc}\quad\quad H_1: \mu_x\ne\mu_{pc}
$$

Выберем уровень значимости $\alpha = 5\%$. Воспользуемся для тестирования $t$-тестом, так как у нас весьма большие выборки, что видно из значений $n_x, n_{pc}$.


```python
df2 = df.dropna(subset=['user_score'])# выкинем NaN'ы
xone_score = df2.query('platform == "XOne"')['user_score']
pc_score = df2.query('platform == "PC"')['user_score']
var_xone = np.var(xone_score, ddof=1)
var_pc = np.var(pc_score, ddof=1)
print(f'var xone = {var_xone:.3f}\n var pc = {var_pc:.3f}')
```

```python
print(f'n_x={xone_score.count()}, n_pc={pc_score.count()}\n')
print(f'xone_mean={xone_score.mean():.3f},', f'pc_mean={pc_score.mean():.3f}\n')
stat, p = ttest_ind(xone_score, pc_score, equal_var=False) #сравнение дисперсий показывает разные дисперсии
print('stat=%.3f, p=%.3f' % (stat, p))
if p > ALPHA:
	print('Нет данных, чтобы отвергнуть нулевую гипотезу')
else:
	print('Отвергаем нулевую гипотезу')
```

Данные говорят в пользу того, что **средние рейтинги XOne и PC равны**. 

## Равенство средних пользовательских рейтингов жанров Action и Sports

Сформулируем наши гипотезы:
$$
H_0: \mu_a =\mu_{s}\quad\quad H_1: \mu_a\ne\mu_{s}
$$

Выберем уровень значимости $\alpha = 5\%$. Воспользуемся для тестирования $t$-тестом, так как у нас весьма большие выборки, что видно из значений $n_a, n_{s}$.

```python
df.genre.unique()
```

```python
action_score = df2.query('genre == "Action"')['user_score']
sports_score = df2.query('genre == "Sports"')['user_score']
var_action = np.var(action_score, ddof=1)
var_sports = np.var(sports_score, ddof=1)
print(f'var action = {var_action:.3f}\n var sports = {var_sports:.3f}')
```

```python
print(f'n_a={action_score.count()}, n_s={sports_score.count()}\n')
print(f'action_mean={action_score.mean():.3f},', f'sports_mean={sports_score.mean():.3f}\n')
stat, p = ttest_ind(action_score, sports_score, equal_var=False)#сравнение дисперсий показывает разные дисперсии
print('stat={0:.3f}, p={1}'.format(stat, p))
if p > ALPHA:
	print('Нет данных, чтобы отвергнуть нулевую гипотезу')
else:
	print('Отвергаем нулевую гипотезу')
```

Данные свидетельствуют в пользу гипотезы о том, что **средние пользовательские рейтинги для жанров _Action_ и _Sports_ все-таки разные**. Скорее всего оценки экшнов выше.
